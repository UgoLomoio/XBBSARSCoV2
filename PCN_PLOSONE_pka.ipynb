```python
spikes = []
for variant in variants:
  for conformation in conformations:
    spikes.append(variant+"_"+conformation)
spikes
```

```python
import pymol
from pymol import cmd

for variant in spikes:

    conformation = variant.split("_")[1]

    for domain, residues in domains.items():
        cmd.do("delete all")

        pdb_file = pdb_dir + conformation + "/" + variant + ".pdb"
        cmd.do("load {}".format(pdb_file))
        print(pdb_file)

        cmd.do("remove hetatm")
        cmd.do("hide licorice")
        residues_str = ""
        for residue in residues:
            residues_str+= str(residue)+"+"
        residues_str = residues_str[:-1]

        #line="color green, (chain "+ chains + ")"
        #cmd.do(line)
        #cmd.do("set cartoon_transparency, 0.5, (chain "+ chains + ")")
        select = "sele resi {} and chain A+B+C".format(residues_str)
        cmd.do(select)
        remove = "remove not sele"
        cmd.do(remove)
        cmd.do("remove chain B+C")
        cmd.do("save {}/Figures/EP/{}/{}_{}.pdb".format(cwd, domain, variant, domain))
```

```python
pka_domain_dir = cwd+"Data/Pka/"
all_extracted_lines = {}
for pka_file in os.listdir(pka_domain_dir):
    if pka_file[-4:] != ".pka":
      continue
    with open(pka_domain_dir+pka_file) as file:
        variant, conformation, domain = pka_file.split("_")
        domain = domain[:-4]
        print(variant, conformation, domain)
        lines = file.readlines()
        extract_lines = []
        started = False
        continue_ = False
        for line in lines:
            if "    pH  unfolded  folded" in line:
                started = True
                continue_ = True
                first = True
            if "The pI is " in line:
                if started:
                    continue_ = False
                    break
            if continue_:
                if not first:
                    splitted_line = [float(line[:7].strip()), float(line[7:16].strip()), float(line[16:].strip())]
                    extract_lines.append([splitted_line[0], splitted_line[1], splitted_line[2]])
                else:
                    first = False
        all_extracted_lines[variant+"_"+conformation+"_"+domain] = extract_lines
```

```python
domains = {
"NTD": np.arange(14, 305, 1),
"RBD":  np.arange(331, 527, 1)
}
```

```python
results = {}

for variant in spikes:
    variant, conformation = variant.split("_")
    if conformation not in results.keys():
        results[conformation] = {}
    if variant not in results[conformation].keys():
        results[conformation][variant] = {}
    for domain, residues in domains.items():
        #print(domain)
        if domain not in results[conformation][variant].keys():
            results[conformation][variant][domain] = {}
            to_extract = "{}_{}_{}".format(variant, conformation, domain)
            print(variant, domain)
            for ph, unfolded, folded in all_extracted_lines[to_extract]:
                #if str(ph)[-2:] == ".0":
                if ph == 7.0:
                    results[conformation][variant][domain]["folded"] = folded
                    results[conformation][variant][domain]["unfolded"] = unfolded


print(results)
```

```python
for conformation, dvariants in results.items():
    print("Confomation: ", conformation)
    for variant, ddomains in dvariants.items():
      print("Variant: ", variant)
      for domain, ep in ddomains.items():
        print("EP{}: {}".format(domain.lower(), ep))
    print("\n")
```

```python
import pandas as pd
dfs = {}

for conformation in conformations:
  df = pd.DataFrame(results[conformation])
  reform = {(outerKey, innerKey): values for outerKey, innerDict in results[conformation].items() for innerKey, values in innerDict.items()}
  df = pd.DataFrame(reform).T
  dfs[conformation] = df
  df.to_csv(cwd+"Data/variants_{}_ep.csv".format(conformation))
dfs["closed"]
```
